```python
import os
import datetime
import random

print("--- Intermediate Python Core Exercise ---")
print("This exercise covers Python fundamentals, data structures, functions,")
print("core libraries (os, datetime, random), file handling, and exception handling.")
print("Estimated completion time: 3-4 hours.")
print("------------------------------------------")

# --- Global Data for the Exercise (Simulated Data) ---
# We'll imagine this is sensor data from various devices.

DEVICE_TYPES = [
    'Temperature_Sensor',
    'Humidity_Sensor',
    'Pressure_Sensor',
    'Light_Sensor'
]

LOCATIONS = [
    'Lab1',
    'Lab2',
    'OfficeA',
    'OfficeB',
    'Warehouse'
]

READING_RANGES = {
    'Temperature_Sensor': (18.0, 30.0),
    'Humidity_Sensor': (30.0, 80.0),
    'Pressure_Sensor': (980.0, 1030.0),
    'Light_Sensor': (100, 1000)
}

print("\n--- Task 1: Data Generation and Basic Structures ---")

raw_sensor_data = []

for i in range(1, 101):

    timestamp_str = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    device_id = f"DEV{i:03d}"

    device_type = random.choice(DEVICE_TYPES)

    location = random.choice(LOCATIONS)

    min_value, max_value = READING_RANGES[device_type]

    value = round(random.uniform(min_value, max_value), 2)

    reading = (
        timestamp_str,
        device_id,
        device_type,
        location,
        value
    )

    raw_sensor_data.append(reading)

print("\nTotal Sensor Readings:", len(raw_sensor_data))

print("\nFirst 5 Sensor Readings:")

for reading in raw_sensor_data[:5]:
    print(reading)
```

```python

temperature_readings = []

for reading in raw_sensor_data:
    if reading[2] == "Temperature_Sensor":
        temp_dict = {
            "timestamp": reading[0],
            "device_id": reading[1],
            "location": reading[3],
            "temperature_celsius": reading[4]
        }
        temperature_readings.append(temp_dict)

print("\nTemperature Readings:")
for temp in temperature_readings:
    print(temp)
```

```python
unique_locations = set()

for reading in raw_sensor_data:
    unique_locations.add(reading[3])

print("\nUnique Locations:")
print(unique_locations)
```

```python
device_stats = {}

for device in DEVICE_TYPES:
    device_stats[device] = (float('inf'), float('-inf'))

for reading in raw_sensor_data:
    device_type = reading[2]
    value = reading[4]

    min_value, max_value = device_stats[device_type]

    if value < min_value:
        min_value = value

    if value > max_value:
        max_value = value

    device_stats[device_type] = (min_value, max_value)

print("\nDevice Statistics:")
for device, stats in device_stats.items():
    print(device, ":", stats)
```

```python

def get_readings_by_location(readings_list, target_location):
    result = []
    for reading in readings_list:
        if reading[2] == target_location:
            result.append(reading)
    return result

raw_sensor_data = [
    ("2025-07-20 09:00:00", "D101", "Hyderabad", 28.5),
    ("2025-07-20 09:10:00", "D102", "Vijayawada", 30.2),
    ("2025-07-20 09:20:00", "D103", "Hyderabad", 29.1),
    ("2025-07-20 09:30:00", "D104", "Guntur", 31.0)
]

result = get_readings_by_location(raw_sensor_data, "Hyderabad")
print(result)
```

```python
def calculate_average_temperature(temp_data_list):
    if len(temp_data_list) == 0:
        return 0.0

    total = 0
    for reading in temp_data_list:
        total += reading["temperature_celsius"]

    return total / len(temp_data_list)

temperature_readings = [
    {
        "timestamp": "2025-07-20 09:00:00",
        "device_id": "D101",
        "location": "Hyderabad",
        "temperature_celsius": 28.5
    },
    {
        "timestamp": "2025-07-20 09:10:00",
        "device_id": "D102",
        "location": "Vijayawada",
        "temperature_celsius": 30.2
    },
    {
        "timestamp": "2025-07-20 09:20:00",
        "device_id": "D103",
        "location": "Hyderabad",
        "temperature_celsius": 29.1
    }
]

average = calculate_average_temperature(temperature_readings)
print("Average Temperature:", average)
```

```python
def generate_report_line(reading_dict):
    return f"[{reading_dict['timestamp']}] Device {reading_dict['device_id']} at {reading_dict['location']}: {reading_dict['temperature_celsius']}°C"

reading = {
    "timestamp": "2025-07-20 09:00:00",
    "device_id": "D101",
    "location": "Hyderabad",
    "temperature_celsius": 28.5
}

report = generate_report_line(reading)
print(report)
```

```python
import datetime

raw_sensor_data = [
    ("2025-07-20 09:00:00", "D101", "Hyderabad", 28.5),
    ("2025-07-20 10:30:00", "D102", "Vijayawada", 30.2),
    ("2025-07-20 12:15:00", "D103", "Hyderabad", 29.1),
    ("2025-07-20 15:45:00", "D104", "Guntur", 31.0)
]

if len(raw_sensor_data) > 1:

    first_timestamp_str = raw_sensor_data[0][0]
    last_timestamp_str = raw_sensor_data[-1][0]

    date_format = "%Y-%m-%d %H:%M:%S"

    first_dt = datetime.datetime.strptime(first_timestamp_str, date_format)
    last_dt = datetime.datetime.strptime(last_timestamp_str, date_format)

    duration = last_dt - first_dt

    days = duration.days
    seconds = duration.total_seconds()

    hours = int(seconds // 3600) % 24
    minutes = int((seconds % 3600) // 60)

    print("Duration:", days, "days,", hours, "hours,", minutes, "minutes")

else:
    print("Not enough data.")
```
